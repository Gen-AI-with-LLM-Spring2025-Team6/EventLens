```python
import requests
import pandas as pd

# API Base URL and Endpoint
root = "https://ensembledata.com/apis"
endpoint = "/instagram/user/posts"

# API Parameters
params = {
    "user_id": 67393179753,
    "depth": 1,
    "oldest_timestamp": 1666262030,
    "chunk_size": 10,
    "start_cursor": "",
    "alternative_method": False,
    "token": ""
}

# Send API request
res = requests.get(root + endpoint, params=params)


if res.status_code == 200:
    data = res.json()

    posts = data.get("data", {}).get("posts", [])
    event_list = []

    for idx, post in enumerate(posts):
        text = "No Text Available"
        text_data = post.get("node", {}).get("edge_media_to_caption", {}).get("edges", [])

        if isinstance(text_data, list) and len(text_data) > 0:
            text = text_data[0].get("node", {}).get("text", "No Text Available")

        video_url = post.get("node", {}).get("video_url", "No Video Found")

        event_list.append({"Post Number": idx + 1, "Text": text, "Video URL": video_url})

    df_insta_events = pd.DataFrame(event_list)

else:
    print(f"Failed to fetch data. Status Code: {res.status_code}")
```

### Method 2 : Using APIFY API

#### Apify Client - Using Actors to scrape posts from Instagram from 4 different pages

```python
from apify_client import ApifyClient

# Initialize the ApifyClient with your API token
client = ApifyClient("")

# Prepare the Actor input
run_input = {
    "directUrls": ["https://www.instagram.com/bostontoday/","https://www.instagram.com/bostondesievents/","https://www.instagram.com/bostoneventguide/","https://www.instagram.com/bostonparksdept/"],
    "resultsType": "posts",
    "resultsLimit": 20,
    "searchType": "hashtag",
    "searchLimit": 1,
    "addParentData": False,
}

# Run the Actor and wait for it to finish
run = client.actor("shu8hvrXbJbY3Eb9W").call(run_input=run_input)
```

```python
data_list = []
for item in client.dataset(run["defaultDatasetId"]).iterate_items():
    data_list.append(item)

df_instagram = pd.DataFrame(data_list)
```

#### Loading Data Into Snowflake Tables

```python
SNOWFLAKE_ACCOUNT=''
SNOWFLAKE_USER=''
SNOWFLAKE_PASSWORD=''
SNOWFLAKE_DATABASE=''
SNOWFLAKE_SCHEMA=''
SNOWFLAKE_WAREHOUSE=''
SNOWFLAKE_ROLE=''
```

```python
import snowflake.connector


conn = snowflake.connector.connect(
    user=SNOWFLAKE_USER,
    password=SNOWFLAKE_PASSWORD,
    account=SNOWFLAKE_ACCOUNT, 
    warehouse=SNOWFLAKE_WAREHOUSE,
    database=SNOWFLAKE_DATABASE,
    schema=SNOWFLAKE_SCHEMA,
    role=SNOWFLAKE_ROLE,
    client_session_keep_alive=True
)


cursor = conn.cursor()
```

```python
def clean_dataframe(df):
    df = df.copy()
    for col in df.columns:
        df.loc[:, col] = df[col].apply(lambda x: str(x) if isinstance(x, (dict, list)) else x) 
        df.loc[:, col] = df[col].apply(lambda x: None if pd.isna(x) else x)
    return df
```

```python
df_instagram_selected_clean = clean_dataframe(df_instagram[[
    "inputUrl", "id", "type", "caption", "hashtags", "url", 
    "displayUrl", "images", "timestamp", "ownerUsername", 
    "videoUrl", "productType", "locationName"
]])
```

```python
data_to_insert = [tuple(row) for row in df_instagram_selected_clean.itertuples(index=False, name=None)]
```

```python
insert_query = """
INSERT INTO BOSTON_EVENTS_INSTAGRAM_DATA (
    inputUrl, id, type, caption, hashtags, url, displayUrl, images, timestamp, 
    ownerUsername, videoUrl, productType, locationName
) VALUES (
    %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s
);
"""
```

```python
cursor.executemany(insert_query, data_to_insert)

conn.commit()
print(f"Inserted {len(data_to_insert)} records into Snowflake!")
```

```python
cursor.close()
conn.close()
```
